```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Create a TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features=1000)  # Limit the number of features for faster computation

# Fit and transform titles and reviews into TF-IDF vectors
title_tfidf = tfidf.fit_transform(df_reviews['title'])
text_tfidf = tfidf.fit_transform(df_reviews['text'])

# Convert to dense matrix to see the features
title_tfidf_dense = title_tfidf.toarray()
text_tfidf_dense = text_tfidf.toarray()

# Add the resulting vectors as new columns to the dataframe (you can use the first few features or all)
df_reviews['title_tfidf'] = list(title_tfidf_dense)
df_reviews['text_tfidf'] = list(text_tfidf_dense)

# Check if the embeddings are added correctly
print(df_reviews[['title', 'title_tfidf', 'text', 'text_tfidf']].head(15))
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Load pre-trained Sentence Transformer model (You can use other models too)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Function to generate embeddings for titles and reviews
def get_embeddings(texts):
    return model.encode(texts, convert_to_tensor=True)

# Example: Create embeddings for product titles and reviews
df_reviews['title_embedding'] = df_reviews['title'].apply(lambda x: get_embeddings([x])[0])  # List of embeddings for titles
df_reviews['text_embedding'] = df_reviews['text'].apply(lambda x: get_embeddings([x])[0])  # List of embeddings for reviews
```

```python
# Make sure user_id and asin are properly indexed for embedding lookups
df_reviews['user_id'] = df_reviews['user_id'].astype('category').cat.codes
df_reviews['asin'] = df_reviews['asin'].astype('category').cat.codes

# Prepare the inputs for training
X_train = df_reviews[['user_id', 'asin']]  # User and Product (ASIN)
y_train = df_reviews['rating']  # Target variable
```

```python
# Flatten the embeddings and concatenate them with user_id and asin
X_train_title_text = np.hstack([np.array(df_reviews['title_tfidf'].tolist()), np.array(df_reviews['text_tfidf'].tolist())])

# Combine all features (user_id, asin, title_embedding, text_embedding)
X_train_full = np.hstack([X_train.values, X_train_title_text])

# Check the shape of your input features
print(X_train_full.shape)  # Ensure it includes user_id, asin, and the embeddings
```

```python
import tensorflow as tf

def DeepFM_model(n_user: int, n_item: int, text_feature_dim: int, dim=8, layers=[16, 16, 16], l2=1e-6) -> tf.keras.Model:
    l2 = tf.keras.regularizers.l2(l2)

    # Inputs
    user_id_input = tf.keras.Input(shape=(), name='user_id', dtype=tf.int32)
    item_id_input = tf.keras.Input(shape=(), name='item_id', dtype=tf.int32)
    text_features_input = tf.keras.Input(shape=(text_feature_dim,), name='text_features', dtype=tf.float32)

    # Embedding layers
    user_embedding = tf.keras.layers.Embedding(n_user, dim, embeddings_regularizer=l2)(user_id_input)
    item_embedding = tf.keras.layers.Embedding(n_item, dim, embeddings_regularizer=l2)(item_id_input)

    # FM interaction
    user_bias = tf.keras.layers.Embedding(n_user, 1, embeddings_initializer='zeros')(user_id_input)
    item_bias = tf.keras.layers.Embedding(n_item, 1, embeddings_initializer='zeros')(item_id_input)
    fm_interaction = tf.reduce_sum(user_embedding * item_embedding, axis=1, keepdims=True) + user_bias + item_bias

    # Deep component
    user_item_concat = tf.keras.layers.Concatenate(axis=1)([user_embedding, item_embedding])
    user_item_flat = tf.keras.layers.Flatten()(user_item_concat)
    deep_input = tf.keras.layers.Concatenate(axis=1)([user_item_flat, text_features_input])

    deep = deep_input
    for layer in layers:
        deep = tf.keras.layers.Dense(layer, activation='relu', kernel_regularizer=l2)(deep)
    deep_output = tf.keras.layers.Dense(1, kernel_regularizer=l2)(deep)

    # Combine FM and Deep output
    output = tf.keras.activations.sigmoid(fm_interaction + deep_output)

    model = tf.keras.Model(inputs=[user_id_input, item_id_input, text_features_input], outputs=output)
    return model
```

```python
# Example dimensions
n_user = df_reviews['user_id'].nunique()
n_item = df_reviews['asin'].nunique()
text_feature_dim = X_train_title_text.shape[1]

# Prepare model
model = DeepFM_model(n_user=n_user, n_item=n_item, text_feature_dim=text_feature_dim)

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])

# Prepare inputs
X_user = df_reviews['user_id'].astype('int32').values
X_item = df_reviews['asin'].astype('int32').values
X_text_features = np.hstack([
    np.array(df_reviews['title_tfidf'].tolist()),
    np.array(df_reviews['text_tfidf'].tolist())
])

# Labels (e.g., 1 if rating ≥ 4, else 0)
y = (df_reviews['rating'] >= 4).astype('int')

# Train-test split
from sklearn.model_selection import train_test_split
X_user_train, X_user_val, X_item_train, X_item_val, X_text_train, X_text_val, y_train, y_val = train_test_split(
    X_user, X_item, X_text_features, y, test_size=0.2, random_state=42
)

# Train
model.fit(
    x=[X_user_train, X_item_train, X_text_train],
    y=y_train,
    validation_data=([X_user_val, X_item_val, X_text_val], y_val),
    epochs=10,
    batch_size=256
)
```

```python
X_test_np = X_test.values  # or X_test.to_numpy()

X_test_inputs = [
    X_test_np[:, 0].astype("int32"),  # user_id
    X_test_np[:, 1].astype("int32"),  # item_id
    X_test_np[:, 2:]                  # tf-idf features
]

test_results = model.evaluate(X_test_inputs, y_test, batch_size=64)
print(f"Test Loss: {test_results[0]:.4f}, Test AUC: {test_results[1]:.4f}")
```
